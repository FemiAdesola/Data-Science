```python
# For making massive paragraph or sentences
sentences = whatsaup_df['Messages'].tolist()
```

```python
sentences
```

```python
len(sentences)
```

```python
# For joining the sentences
sentences_as_one_string = " ".join(map(str, sentences))
```

```python
sentences_as_one_string
```

```python
from wordcloud import WordCloud
import nltk

nltk.download('stopwords')
```

```python
w = WordCloud()
stop_words = list(w.stopwords)
custom_stop_words = ['Jesus', 'brother', 'Happy','enemy', 'wife', 'cut', 'FM', 'days', 'family', 'man','children','hot', 'pick','sick', 'sake', 'deads','done','things', 'done', 'sister', 'Evelyn', 'us','please', 'dad', 'fall', 'pastor', 'labour', 'everything', 'thee', 'christ', 'mighty', 'land', 'thing', 'Almighty', 'name', 'birthday', 'many','every','Amen' ,'long', 'bring', 'uncle', 'bro', 'celebrate', 'new', 'age', 'thank', 'wishing', 'God', 'Femi', 'prayer', 'Sir', 'Lord', 'May', 'year', 'life', 'Jesu', 'Will', 'day', 'power', 'years']
stop_words = set(stop_words + custom_stop_words)

wordcloud = WordCloud(background_color="black", width=1600,height=800, stopwords=stop_words).generate(sentences_as_one_string)
plt.figure(figsize=(20,10), facecolor='k')
plt.imshow(wordcloud)
plt.axis('off')
```
